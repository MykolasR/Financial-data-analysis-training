```python
import pandas as pd
import matplotlib.pyplot as plt

# matplotlib move to notebook format

%matplotlib notebook
%matplotlib inline
```

```python
data = pd.read_csv("C:/Users/Vartotojas/Desktop/Financial analysis training/AAPL.csv", index_col = 0, parse_dates =True)
```

```python
%matplotlib inline
data.plot()
```

```python
data["Close"].plot()
```

```python
# it means we create new figure "fig" with axis of figure "ax". You have to make a new figure bacause matplotlib not understand that we work on different plot
fig, ax = plt.subplots()
data["Close"].plot(ax=ax)

ax.set_ylabel("Price")
ax.set_title("AAPL")
```

```python
#create 4 dataframes with title and tight layout ensure clear view
fig, ax = plt.subplots(2,2)

data["Open"].plot(ax=ax[0,0], title="Open")
data["High"].plot(ax=ax[0,1], title="High")
data["Low"].plot(ax=ax[1,0], title="Low")
data["Close"].plot(ax=ax[1,1], title="Close")

plt.tight_layout()
```

```python
#indexing during a plot, barh rotate bars to ensure better view
fig, ax = plt.subplots()
data["Volume"].loc["2022-07-01": "2022-08-01"].plot.barh(ax=ax)
```
